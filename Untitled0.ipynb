```python
import numpy
import csv

def getData():
  size = []
  avgTime = []
  with open('size.csv', newline="") as f:
    filedata = csv.DictReader(f)
    for val in filedata:
      size.append(float(val['Size of TV']))
      avgTime.append(float(val['\tAverage time spent watching TV in a week (hours)']))
  return {"x":size, "y":avgTime}

def getVal(data):
  cor = numpy.corrcoef(data['x'], data['y'])
  print('Value: ',cor[0,1])

e = getData()
getVal(e)
```
